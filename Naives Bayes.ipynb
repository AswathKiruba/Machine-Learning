Naive Bayes methods are a set of supervised learning algorithms based on applying Bayes’ theorem with the “naive” assumption of independence between every pair of features. Given a class variable y and a dependent feature vector x1 through xn, Bayes’ theorem states the following relationship:

    P(y∣x1,…,xn)=P(y)P(x1,…xn∣y)P(x1,…,xn)
    P(y∣x1,…,xn)=P(y)P(x1,…xn∣y)P(x1,…,xn)
Using the naive independence assumption that
        P(xi|y,x1,…,xi−1,xi+1,…,xn)=P(xi|y)
        P(xi|y,x1,…,xi−1,xi+1,…,xn)=P(xi|y)
for all i, this relationship is simplified to:

    P(y∣x1,…,xn)=P(y)∏ni=1P(xi∣y)P(x1,…,xn)
    P(y∣x1,…,xn)=P(y)∏i=1nP(xi∣y)P(x1,…,xn)
We now have a relationship between the target and the features using Bayes Theorem along with a Naive Assumption that all features are independent.

```python
nyc_sub = pd.read_csv('C:/Users/aswat/nyc_subway_weather.csv')
```

```python
df1 = nyc_sub.copy()
#drop rows with NA
df1.dropna(inplace = True)
df1.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
nyc_data = df1[['fog','precipi','pressurei','rain','tempi','wspdi']]
```

```python
nyc_target = df1[['conds']]
```

```python
nyc_complete = pd.concat([nyc_data,nyc_target],axis=1)
```

```python
nyc_complete.head()
```

```python
X = nyc_data
Y = nyc_target
Y = np.ravel(Y)
```

```python
#Fit a naives Bayes model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
from sklearn.cross_validation import train_test_split

# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
# Fit the training model
model.fit(X_train,Y_train)
```

```python
# Predicted outcomes
predicted = model.predict(X_test)

# Actual Expected Outvomes
expected = Y_test
```

```python
print(metrics.accuracy_score(expected, predicted))
```

The accuracy is low when compared to that of the KNN
